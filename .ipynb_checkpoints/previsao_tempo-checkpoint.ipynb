#### Importação de Bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from os import listdir
from os.path import isfile, join
```

```python
#Lista arquivos disponíveis 
files = [f for f in listdir(os.getcwd()) if isfile(join(os.getcwd(), f))]
files
```

```python
# Criando um Dataframe vazio para receber todos dos dados
# Create a empty Dataframe to store all data
df = pd.DataFrame()  
```

```python
%%time
# Lista para nomes das colunas, necessário retirar caracteres
# List of columns names, needed to extract all missing characters
renomear_colunas = ["DATA (YYYY-MM-DD)","HORA (UTC)",
                    "PRECIPITACAO TOTAL, HORARIO (mm)",
                    "PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)",
                    "PRESSAO ATMOSFERICA MAX.NA HORA ANT.(mB)",
                    "PRESSAO ATMOSFERICA MIN. NA HORA ANT.(mB)",
                    "RADIACAO GLOBAL (W/m2)",
                    "TEMPERATURA DO AR - BULBO SECO, HORARIA (C)",
                    "TEMPERATURA DO PONTO DE ORVALHO (C)",
                    "TEMPERATURA MAXIMA NA HORA ANT.(C)",
                    "TEMPERATURA MINIMA NA HORA ANT.(C)",
                    "TEMPERATURA ORVALHO MAX. NA HORA ANT.(C)",
                    "TEMPERATURA ORVALHO MIN. NA HORA ANT.(C)",
                    "UMIDADE REL. MAX. NA HORA ANT.(%)",
                    "UMIDADE REL. MIN. NA HORA ANT.(%)",
                    "UMIDADE RELATIVA DO AR, HORARIA(%)",
                    "VENTO, DIRECAO HORARIA (gr)",
                    "VENTO, RAJADA MAXIMA (m/s)",
                    "VENTO, VELOCIDADE HORARIA (m/s)",
                    "Unnamed"]
# O ultimo item da lista, serve para capturar os dados que serão importados  já que cada linha dos arquivos terminam com ;
# The last item of the list is needed to catch all data, there is a wrong format in the source file
```

```python
%%time
# Percorre a lista de arquivos csv e fazendo o append no DataFrame
# Run through the file list, storing all data in the Dataframe
for i in files:
    if(i[-3:])=="CSV":
        df = df.append(pd.read_csv(i,sep=";", encoding="cp860", skiprows=8, names = renomear_colunas, header = 0),ignore_index = True)
```

```python
# Excluindo coluna de lixo
# Deleting the last column
df.drop("Unnamed", axis=1, inplace = True)
```

```python
df.columns
```

```python
df.describe()
```

```python
%%time
# Altera , para . para converter de str para float
# Replace , to . to convert from str to float
df.replace(",",".", regex = True, inplace = True)
```

```python
%%time
# Altera / para - para converter de str para datetime
# Replace / to - to convert from str to datetime
df["DATA (YYYY-MM-DD)"].replace("/","-", regex = True, inplace = True)
```

```python
%%time
# Uniformizando as horas
# Correct the hour format
df["HORA (UTC)"].replace(" UTC","", regex = True, inplace = True)
```

```python
# Criando coluna com Data e Hora para criação do tipo Datetime
df["DATA_HORA"] = df["DATA (YYYY-MM-DD)"] +" " + df["HORA (UTC)"]
```

```python
%%time
# convertando para datetime
# Convert to Datetime
df["DATA_HORA"] = pd.to_datetime(df["DATA_HORA"], errors = "ignore", format="%Y-%m-%d %H:%M")
```

```python
# Ordenando
# Sorting
df.sort_values(by=["DATA_HORA"], inplace = True)
```

```python
# Verificando a ordenação das datas
# Checking 

df['DATA_HORA']
```

```python
%%time
# Altera os dados -9999 para NaN para uma melhor estratégia de substituição dos dados faltantes
# Replace -9999 to Nan
df.replace("-9999.000000", np.NaN, inplace=True)
df.replace("-9999", np.NaN, inplace=True)
df.replace(-9999, np.NaN, inplace=True)
```

```python
# Reset do index
df.reset_index(drop = True)
# Quando o reset_index é usado, o indice antigo é adiconado ao Dataframe como uma coluna, portanto precisa ser excluido utilizando o parametro drop = True
```

```python
df["PRECIPITACAO TOTAL, HORARIO (mm)"] = pd.to_numeric(df["PRECIPITACAO TOTAL, HORARIO (mm)"])

df["PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)"] = pd.to_numeric(df["PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)"])

df["PRESSAO ATMOSFERICA MAX.NA HORA ANT.(mB)"] = pd.to_numeric(df["PRESSAO ATMOSFERICA MAX.NA HORA ANT.(mB)"])

df["PRESSAO ATMOSFERICA MIN. NA HORA ANT.(mB)"] = pd.to_numeric(df["PRESSAO ATMOSFERICA MIN. NA HORA ANT.(mB)"])

df["RADIACAO GLOBAL (W/m2)"] = pd.to_numeric(df["RADIACAO GLOBAL (W/m2)"])

df["TEMPERATURA DO AR - BULBO SECO, HORARIA (C)"] = pd.to_numeric(df["TEMPERATURA DO AR - BULBO SECO, HORARIA (C)"])

df["TEMPERATURA DO PONTO DE ORVALHO (C)"] = pd.to_numeric(df["TEMPERATURA DO PONTO DE ORVALHO (C)"])

df["TEMPERATURA MAXIMA NA HORA ANT.(C)"] = pd.to_numeric(df["TEMPERATURA MAXIMA NA HORA ANT.(C)"])

df["TEMPERATURA MINIMA NA HORA ANT.(C)"] = pd.to_numeric(df["TEMPERATURA MINIMA NA HORA ANT.(C)"])

df["TEMPERATURA ORVALHO MAX. NA HORA ANT.(C)"] = pd.to_numeric(df["TEMPERATURA ORVALHO MAX. NA HORA ANT.(C)"])

df["TEMPERATURA ORVALHO MIN. NA HORA ANT.(C)"] = pd.to_numeric(df["TEMPERATURA ORVALHO MIN. NA HORA ANT.(C)"])

df["VENTO, RAJADA MAXIMA (m/s)"] = pd.to_numeric(df["VENTO, RAJADA MAXIMA (m/s)"])

df["VENTO, VELOCIDADE HORARIA (m/s)"] = pd.to_numeric(df["VENTO, VELOCIDADE HORARIA (m/s)"])
```
